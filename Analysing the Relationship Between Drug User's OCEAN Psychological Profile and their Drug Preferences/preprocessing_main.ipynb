### Transform Dataset from .data to .csv & clean data if necessary

```python
column_head = ["ID","Age","Gender","Education","Country","Ethnicity","NScore","EScore","OScore","AScore","CScore","Impulsive","SS","Alcohol","Amphet","Amyl","Benzos","Caff","Canab","Choc","Coke","Crack","Ecstasy","Heroin","Ket","LegalH","LSD","Meth","Mushroom","Nicotine","Semer","VSA"]
row_1 = pre_df.columns.to_list()
#column head is the first row therefore append column head to first row of dataset
nrow_1 = []
for i in row_1:
    if "C" not in i:
        nrow_1.append(float(i))
    else:
        nrow_1.append(i)
# Add nrow_1 into df
pre_df.loc[-1] = nrow_1
pre_df.index = pre_df.index + 1
#Replace column headers
pre_df.columns = column_head
#Sort values by ID
post_df = pre_df.sort_values(by="ID")
post_df.head(5)
```

## Group Drugs into based on CNS  Effect Classification
* Depressants: Benzos, Heroin, Methadone, Amyl Nitrate
* Stimulants: Amphetamines, Ecstasy, Cocaine, Crack
* Psychedelics: Mushrooms, LSD, Ketamine
* Excluded: Alcohol, Cannabis, Caffeine, Nicotine, Chocolate, Legal Highs, Volatile Substance Abuse
* Legal Highs and Volatile substance abuse were removed as they do not fall under either category
* Cannabis, Caffeine,Alcohol and Chocolate are commonly consumed drugs therefore including them in the different drug categories would result in a high degree of overlap between the drug categories

Higher frequency usage is a good indication of drug preferences.
* Users: CL4,CL5,CL6
* Non-Users: CL0,CL1,CL2,CL3


#### Depressants

```python
user = ["CL4","CL5","CL6"]
non = ["CL0","CL1","CL2","CL3"]
depress = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([benz in user,he in user, met in user, amy in user]) & all([am in non, ec in non, co in non,cr in non,ls in non,k in non,mush in non]):
        depress.append(1)
    else:
        depress.append(0)
sum(depress)
```

#### Stimulants

```python
user = ["CL4","CL5","CL6"]
non = ["CL0","CL1","CL2","CL3"]
stim = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([am in user, ec in user, co in user,cr in user]) & all([benz in non,he in non, met in non, amy in non,ls in non,k in non,mush in non]):
        stim.append(1)
    else:
        stim.append(0)
sum(stim)
```

#### Hallucinogens

```python
user = ["CL4","CL5","CL6"]
non = ["CL0","CL1","CL2","CL3"]
hallu = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([ls in user, mush in user, k in user]) & all([benz in non,he in non, met in non, amy in non,am in non, ec in non, co in non,cr in non]):
        hallu.append(1)
    else:
        hallu.append(0)

sum(hallu)
```

#### Calculate similarity between each category
* Use Jaccard Index i.e. Intersection Over Union , higher score = higher overlap

```python
from sklearn.metrics import jaccard_score as js
print(f"Hallu+Depressants:{js(hallu,depress)}")
print(f"Hallu+Stimulants:{js(hallu,stim)}")
print(f"Stimulants+Depressants:{js(stim,depress)}")
```

```python
df["Depress"] = depress
df["Stim"] = stim
df["Hallu"] = hallu
df.head()
```

## Seperate Dataset into Overall Drug User/ Non-Users
* Month based user: CL4,CL5,CL6
* Week based User: CL5,CL6
* Day based User:

#### Monthly Users

```python
user = ["CL4","CL5","CL6"]
non = ["CL0","CL1","CL2","CL3"]
month_u = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([benz in user,he in user, met in user, amy in user,am in user, ec in user, co in user,cr in user,ls in user, mush in user, k in user]):
        month_u.append(1)
    else:
        month_u.append(0)
sum(month_u)
```

#### Weekly Users

```python
user = ["CL5","CL6"]
non = ["CL0","CL1","CL2","CL3","CL4"]
week_u = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([benz in user,he in user, met in user, amy in user,am in user, ec in user, co in user,cr in user,ls in user, mush in user, k in user]):
        week_u.append(1)
    else:
        week_u.append(0)
sum(week_u)
```

#### Daily Users

```python
user = ["CL6"]
non = ["CL0","CL1","CL2","CL3","CL4"]
day_u = []
for am,amy,benz,co,cr,ec,he,k,lh,ls,met,mush,vs in zip(df["Amphet"],df["Amyl"],df["Benzos"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LegalH"],df["LSD"],df["Meth"],df["Mushroom"],df["VSA"]):
    if any([benz in user,he in user, met in user, amy in user,am in user, ec in user, co in user,cr in user,ls in user, mush in user, k in user]):
        day_u.append(1)
    else:
        day_u.append(0)
sum(day_u)
```

```python
print(f"Monthly+Weekly:{js(month_u,week_u)}")
print(f"Monthly+Daily:{js(month_u,day_u)}")
print(f"Daily+Weekly:{js(day_u,week_u)}")
```

High degree of overlap is observed between all categories. As we are trying to determine how users frequency of drug use changes based on their psychological profile, overlaps should not affect our analysis greatly.
* Use monthly users result as a baseline
* compare weekly results to monthly - new baseline for daily
* compare daily results to weekly

```python
df["monthly_u"] = month_u
df["weekly_u"] = week_u
df["daily_u"] = day_u
df.head()
```

## Seperate Users based on Legal Drug Use
* Amphetamines, Benzodiazapines and Methadone are not considered in this analysis as these are drugs that could be prescribed,therefore hard to distinguish between necessities or drug abuse.
* Cannabis is included in the legal drug category as general public preception of cannabis has shifted greatly.
* The point of this comparison is to determine high risk taking behaviour.

Legal Drugs: Alcohol,Canab, Amyl,LegalH,Methadone,Nicotine 

Illegal Coke,Crack,Ecstasy,Heroin,Ket,LSD


```python
usage = ["CL4","CL5","CL6"]
use=["CL0","CL1","CL2","CL3"] # Non-User threshold

legal = []
for a,c,am,lh,meth,nic,vs,c,cr,ec,her,ket,lsd in zip(df["Alcohol"],df["Canab"],df["Amyl"],df["LegalH"],df["Meth"],df["Nicotine"],df["VSA"],df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LSD"]):
    if any([a in usage,c in usage,am in usage,lh in usage,meth in usage,nic in usage,vs in usage]) &all([c in use,cr in use,ec in use,her in use,ket in use,lsd in use]):
        legal.append(1)
    else:
        legal.append(0)

illegal = []

for c,cr,ec,her,ket,lsd in zip(df["Coke"],df["Crack"],df["Ecstasy"],df["Heroin"],df["Ket"],df["LSD"]):
    if any([c in usage,cr in usage,ec in usage,her in usage,ket in usage,lsd in usage]):
        illegal.append(1)
    else:
        illegal.append(0)
        

print(sum(legal))
print(sum(illegal))
js(legal,illegal)
```
