```python
import pandas as pd
from imblearn.over_sampling import SMOTENC
```

```python
df = pd.read_csv("bank-full.csv",sep=";")
len(df)
```

```python
#Determining if class imblanace is present
df.groupby("y").count()
```

```python
#Removing redundant columns day and month
del df["day"]
del df["month"]
```

```python
#Convert strings to labels
jobs = []
for i in df["job"]:
    if i == "unknown":
        jobs.append(-1)
    elif i == "unemployed":
        jobs.append(0)
    elif i == "student":
        jobs.append(1)
    elif i == "blue-collar":
        jobs.append(2)
    elif i == "housemaid":
        jobs.append(3)
    elif i == "technician":
        jobs.append(4)
    elif i == "admin.":
        jobs.append(5)
    elif i == "services":
        jobs.append(6)
    elif i == "self-employed":
        jobs.append(7)
    elif i == "management":
        jobs.append(8)
    elif i == "entrepreneur":
        jobs.append(9)
    elif i == "retired":
        jobs.append(12)
df["job"] = jobs
```

```python
martial = []
for i in df["marital"]:
    if i == "unkown":
        martial.append(-1)
    elif i == "single":
        martial.append(0)
    elif i == "divorced":
        martial.append(1)
    elif i == "married":
        martial.append(2)
df["marital"] = martial
```

```python
education = []
for i in df["education"]:
    if i == "unknown":
        education.append(-1)
    elif i =="primary":
        education.append(0)
    elif i =="secondary":
        education.append(1)
    elif i =="tertiary":
        education.append(2)
df["education"] = education
```

```python
def to_bin(x):
    if x == "yes":
        return 1
    elif x == "no":
        return 0
```

```python
default = []
for i in df["default"]:
     default.append(to_bin(i))
df["default"] = default

house = []
for i in df["housing"]:
    house.append(to_bin(i))
df["housing"] = house

loan = []
for i in df["loan"]:
    loan.append(to_bin(i))
df["loan"] = loan

y = []
for i in df["y"]:
    y.append(to_bin(i))
df["y"] = y
```

```python
df.head()
```

```python
con = []
for i in df["contact"]:
    if i == "unknown":
        con.append(-1)
    elif i == "telephone":
        con.append(0)
    elif i == "cellular":
        con.append(1)
df["contact"] = con
```

```python
pout = []
for i in df["poutcome"]:
    if i == "unknown":
        pout.append(-1)
    elif i == "failure":
        pout.append(0)
    elif i == "other":
        pout.append(1)
    elif i == "success":
        pout.append(2)
df["poutcome"] = pout
```

```python
#Scaling values between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = df[["age","job","marital","education","default","housing","loan","contact","campaign","pdays","previous","poutcome"]]
head = ["age","job","marital","education","default","housing","loan","contact","campaign","pdays","previous","poutcome"]

numerical_new = scaler.fit_transform(df)
numeriacl_pd = pd.DataFrame(numerical_new,columns=df.columns)
numeriacl_pd.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:14]
Y = df.iloc[:,-1]

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=88)
```

```python
x_train
```

```python
#Balancing Classes Using Smote NC
sm = SMOTENC(random_state=88, categorical_features=[1,2,3,4,6,7,8,12]) # Features which are categorical
x_smote,y_smote = sm.fit_resample(x_train,y_train)
```

```python
test = pd.DataFrame(x_smote,columns=x_train.columns)
test
```

```python
test["y"] = y_smote
test.head()
```

```python
test.to_csv("train_balanced.csv",index=False)
```
